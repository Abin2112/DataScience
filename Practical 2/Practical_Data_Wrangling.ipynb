##1. Import Libraries

```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Load the Dataset

```python
# Load the dataset
df = pd.read_csv("shopping_behavior_updated (1).csv")
```

```python
# Show first 5 rows
df.head()
```

## Describing the Data

```python
print("Dimensions: {}". format(df.shape))
```

```python
df.describe()
```

## Navigating the DataFrame

```python
print(df.iloc[0])
```

```python
print(df.iloc[1:4])
```

```python
print(df.iloc[:5])
```

```python
df = df.set_index(df['Item Purchased'])
```

## Set DataFrame Index to a Unique Column

> Add blockquote



```python
df = df.set_index('Customer ID')

df.head()
```

## Selecting Rows Using Conditionals

```python
#select customers above age 30
df[df['Age'] > 30]
```

```python
# Female customers who purchased electronics
df[(df['Gender'] == 'Female') & (df['Item Purchased'] == 'Electronics')]
```

## Replacing Values

```python
# Replace "M" with "Male" and "F" with "Female"
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})
```

## Renaming Columns

```python
# Renaming columns properly
df = df.rename(columns={
    'Age': 'Customer Age',
    'Gender': 'Customer Gender'
})
df.head()
```

## Finding Min, Max, Sum, Average, Count

```python
print('Maximum: {}'.format(df['Customer Age'].max()))
print('Minimum: {}'.format(df['Customer Age'].min()))
print('Mean: {}'.format(df['Customer Age'].mean()))
print('Sum: {}'.format(df['Customer Age'].sum()))
print('Count: {}'.format(df['Customer Age'].count()))
```

```python
print("Variance: {}".format(df['Customer Age'].var()))
print("Standard Deviation: {}".format(df['Customer Age'].std()))
print("Kurtosis: {}".format(df['Customer Age'].kurtosis()))
print("Skewness: {}".format(df['Customer Age'].skew()))
```

## Finding Unique Values

```python
# Unique items purchased
df['Item Purchased'].unique()
```

```python
# Number of unique categories
df['Category'].nunique()
```

## Handling Missing Values

```python
# Show missing values
df.isnull().sum()
```

```python
# Fill missing numeric values with mean
df['Customer Age'] = df['Customer Age'].fillna(df['Customer Age'].mean())
```

```python
# Drop rows with any missing values
df = df.dropna()
```

## Deleting a Column

```python
# Delete column
df = df.drop('Category', axis=1)
df.head()
```

## Deleting a Row

```python
# Delete a row by index (example: CustomerID = 1003)
df = df.drop(1003)
```

## Applying Function Over

---

Column Elements

```python
# Example: applying 18% GST to purchase amount
df['Amount_with_GST'] = df['Purchase Amount (USD)'].apply(lambda x: x * 1.18)
df.head()
```

## Applying a Function to Groups

```python
# Average purchase amount by gender
df.groupby('Customer Gender')['Purchase Amount (USD)'].mean()
```

```python
# Count purchases by item type
df.groupby('Item Purchased')['Purchase Amount (USD)'].count()
```

## Concatenating DataFrames

```python
# Creating two small dfs to concatenate
df1 = df.head(5)
df2 = df.tail(5)

# Concatenate vertically
merged = pd.concat([df1, df2], axis=0)
merged
```

## Plotting

### Bar Chart  Purchase Amount by Gender

```python
df.groupby('Customer Gender')['Purchase Amount (USD)'].mean().plot(kind='bar')

plt.title("Average Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Avg Purchase (USD)")
plt.show()
```

### Histogram  Age Distribution

```python
df['Customer Age'].hist()

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## GROUPING ROWS BY VALUES

```python
# Group by Gender and print each group
grouped = df.groupby("Customer Gender")

for gender, group in grouped:
    print("\n==============================")
    print(f"GROUP: {gender}")
    print("==============================")
    print(group.head())
```

```python
# Average purchase amount by age group
df.groupby("Customer Age")["Purchase Amount (USD)"].mean()
```

```python
# Number of purchases for each Item Purchased
df.groupby("Item Purchased")["Purchase Amount (USD)"].count()
```

## LOOPING OVER A COLUMN

```python
# Loop through all customer ages

for age in df["Customer Age"]:
    print(age)
```

```python
# Loop through Item Purchased column with index
for index, item in df["Item Purchased"].items():
    print(f"Customer ID: {index}, Item: {item}")
```

## LOOPING OVER GROUPS

```python
for gender, group in df.groupby("Customer Gender"):
    print("\n--- Group:", gender, "---")
    print("Number of customers:", len(group))
```
